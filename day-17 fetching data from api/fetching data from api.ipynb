## API - Application programming interface
- Will learn how to fetch json from API and convert it into dataframe.
- Will fetch news from - https://newsapi.org/
- here is the url - https://newsapi.org/v2/everything?q=tesla&from=2024-02-29&sortBy=publishedAt&apiKey=YOUR_API_ID


```python
import pandas as pd
import requests
```

```python
responce = requests.get("https://newsapi.org/v2/top-headlines/sources?apiKey=YOUR_API_ID&language=en")
responce.json()
```

```python
df = pd.DataFrame(responce.json()['sources'])[['name', 'description', 'category']]
df
```
