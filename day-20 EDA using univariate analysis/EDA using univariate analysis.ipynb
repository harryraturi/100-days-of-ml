## Exploratory data analysis - EDA

### Univariate analysis
- When we have to analyze on single variable it's called univariate analysis.

### Data type
- Numeric : Height, Weight, Age etc.
- Categorical : Country, Month, Gender etc.

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.sample(5)
```

```python
df.info()
```

# B. Pi - Chart
- It's show data in percentage(%).

### Distplot also show the skewness
- Skew values:
    - 0 : Perfectly skewed / Normal distribution
    - +ve( >0) : Positive / Right skewed `eg: - income or wealth or salary`
    - -ve(<0) : Negative / Left skewed    `eg: - marks`



```python
df.Age.skew()
```
