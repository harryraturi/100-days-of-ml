## Convertors parameter

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
pd.read_csv('dataset/IPL Matches 2008-2020.csv',converters={'team1':rename})
```
